```julia
include("TriaxSZ.jl")

θ = [deg2rad(v) for v in collect(range(-0.5, 0.5, length = 100))]
z = 0.75
M = 1e13M_sun
model = Battaglia16ThermalSZProfile()
cosmo = cosmology()
z2r_interpolator = build_z2r_interpolator(1e-5, 5., cosmo);
R_rot = RotationEulerMatrix(0.,0.,1π)
ymap = pressure_profile(model, θ, M, z; ϵ₁ = 0.0, ϵ₂ = 0.95,  
        ϕ = 0., θ = 0., ψ = 0., z2r = z2r_interpolator);
heatmap(log10.(abs.(ymap)))
```
